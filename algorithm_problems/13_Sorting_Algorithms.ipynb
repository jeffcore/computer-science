### Bubble Sort
swaps as it compares<br>
compares left to right, over and over again<BR>
on each pass it keeps moving the largest value to the top and then shortens the length of the array for the next pass  
after each step the largest number will be a the top
    
Naive approach

Complexity:  O(n^2) worst and average case,   O(n) best case

space complexity: O(1) great example of inplace sorting algorithms

```python
def bubble_sort(alist):
    # shortens array after every pass, because highest value will be moved to top after every pass
    for passnum in range(len(alist)-1, 0, -1):
        for i in range(passnum):
            this_item = alist[i]
            next_item = alist[i+1]
            if this_item > next_item:
                temp = this_item
                alist[i] = next_item
                alist[i+1] = this_item
                
test_list = [31,13,9,1,15,7,29]                

bubble_sort(test_list)
print(test_list)

# alternative
def bubble_sort_1(l):
    for iteration in range(len(l)):
        for index in range(1, len(l)):
            this = l[index]
            prev = l[index - 1]

            if prev <= this:
                continue

            l[index] = prev
            l[index - 1] = this
            
wakeup_times = [16,49,3,12,56,49,55,22,13,46,19,55,46,13,25,56,9,48,45]
bubble_sort_1(wakeup_times)
print ("Pass" if (wakeup_times[0] == 3) else "Fail")


# highest to lowest
# Entries are (h, m) where h is the hour and m is the minute
sleep_times = [(24,13), (21,55), (23,20), (22,5), (24,23), (21,58), (24,3)]

def bubble_sort_2(l):
    # TODO: Implement bubble sort solution
    for iteration in range(len(l)):
        for index in range(1, len(l)):
            this_hour, this_min = l[index]
            prev_hour, prev_min = l[index - 1]

            if prev_hour > this_hour or (prev_hour == this_hour and prev_min > this_min):
                continue

            l[index] = (prev_hour, prev_min)
            l[index - 1] = (this_hour, this_min)
            
bubble_sort_2(sleep_times)
print ("Pass" if (sleep_times == [(24,23), (24,13), (24,3), (23,20), (22,5), (21,58), (21,55)]) else "Fail")
```

#### Recursive Bubble Sort

```python

def recursive_bubble_sort(testVariable, length):
	# Base case
	if length <= 1 :
		return
	
	# Recursive case
	# Sort first n-1 elements
	recursive_bubble_sort(testVariable, length - 1)

	# Insert last element at its correct position in sorted array
	lastElement = testVariable[length - 1] # fetch the last element
	temp = length - 2 # start finding its correct location from one element before it
	
	# Move elements of testVariable[0..i-1], that are greater than key, to one position ahead of their current position 
	while (temp >= 0 and testVariable[temp] > lastElement):
		testVariable[temp + 1] = testVariable[temp]
		temp = temp - 1

	testVariable[temp + 1] = lastElement # place the element in its correct position


# Driver Code
testVariable = [5, 4, 2, 3, 1]
print("Original Array ---> " + str(testVariable))

recursive_bubble_sort(testVariable, len(testVariable))
print("Modified Array ---> " + str(testVariable))
```

### Insertion Sort
have a sorted group at beginning of array, keep moving in elements from the right<br>
O(n^2)

```python
def insertionSort(alist):
    for index in range(1,len(alist)):

        currentvalue = alist[index]
        position = index

        while position > 0 and alist[position - 1] > currentvalue:
            alist[position] = alist[position - 1]
            position -= 1

        alist[position] = currentvalue

alist = [54,26,93,17,77,31,44,55,20]
insertionSort(alist)
print(alist)
```

### Shell Sort
Complexity: O(n^2) to O(n)<br>
https://www.youtube.com/watch?v=M9YCh-ZeC7Y <BR>
https://www.youtube.com/watch?v=CmPA7zE8mx0<BR>
http://interactivepython.org/runestone/static/pythonds/SortSearch/TheShellSort.html<br>
sorting by gaps - first gap is n/2   next gap is n/2 - 1<BR>
   the other part is if you do swap you have to check the swapped element with the element gap number before it

```python
def shellSort(alist):
    sublistcount = len(alist)//2
    while sublistcount > 0:

      for startposition in range(sublistcount):
        gapInsertionSort(alist,startposition,sublistcount)

      print("After increments of size",sublistcount,
                                   "The list is",alist)

      sublistcount = sublistcount // 2

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap):

        currentvalue = alist[i]
        position = i

        while position>=gap and alist[position-gap]>currentvalue:
            alist[position]=alist[position-gap]
            position = position-gap

        alist[position]=currentvalue

alist = [54,26,93,17,77,31,44,55,20]
shellSort(alist)
print(alist)
```

### Merge Sort
Divide and Conquer

O(n log(n))
space complexity: O(n) - auxilary arrays have to copy arrays in to new arrays

<img src="../images/mergesort.png">

```python
# easier version of mergesort where merged array is returned
def mergesort(items):

    if len(items) <= 1:
        return items
    
    mid = len(items) // 2
    left = items[:mid]
    right = items[mid:]
    
    left = mergesort(left)
    right = mergesort(right)
    
    return merge(left, right)
    
def merge(left, right):
    
    merged = []
    left_index = 0
    right_index = 0
    
    while left_index < len(left) and right_index < len(right):
        if left[left_index] > right[right_index]:
            merged.append(right[right_index])
            right_index += 1
        else:
            merged.append(left[left_index])
            left_index += 1

    merged += left[left_index:]
    merged += right[right_index:]
        
    return merged


test_list_1 = [8, 3, 1, 7, 0, 10, 2]
test_list_2 = [1, 0]
test_list_3 = [97, 98, 99]
print('{} to {}'.format(test_list_1, mergesort(test_list_1)))
print('{} to {}'.format(test_list_2, mergesort(test_list_2)))
print('{} to {}'.format(test_list_3, mergesort(test_list_3)))
```

```python
def quicksort(alist):
    quick_sort_helper(alist, 0, len(alist) - 1)

def quick_sort_helper(alist, first, last):
    # array is changed inplace
    # loop from start to end of array    
    if first < last:
        # run through quicksort comparing and return new pivot point
        split_point = partition(alist, first, last)
        
        # with new pivote point split array and start over again 
        # do not include new pivot item
        quick_sort_helper(alist, first, split_point - 1)
        quick_sort_helper(alist, split_point + 1, last)

def partition(alist, first, last):
    # this does all of the comparisons against pivot
    
    # get new pivot point always first value
    pivot_value = alist[first]
    print(f'pvitoe value {pivot_value}')

    left_mark = first + 1  # WHY???? because we always grab first element as pivot point or get median of first and last element
    right_mark = last
    
    while True:
        # 
        # compare left side of array to pivot value
        # check you do not go into right area
        # if left mark smaller stop loop
        while left_mark <= right_mark and alist[left_mark] <= pivot_value:
            left_mark = left_mark + 1
            
        # compare right side of array to pivot value
        # if right mark is larget stop loop
        while alist[right_mark] >= pivot_value and right_mark >= left_mark:
            right_mark = right_mark - 1
            
        # now we have left and right mark to do swap
        if right_mark < left_mark:            
            break
        else:
            temp = alist[left_mark]
            alist[left_mark] = alist[right_mark]
            alist[right_mark] = temp
        
    # move pivot to new splitpoint start over    
    temp = alist[first]
    alist[first] = alist[right_mark]
    alist[right_mark] = temp
    print(alist)
    
    return right_mark


test = [21, 4, 1, 3, 9, 20, 25, 6, 5, 14]
print(quicksort(test))
print(test)
```

```python
# This function takes last element as pivot, places 
# the pivot element at its correct position in sorted 
# array, and places all smaller (smaller than pivot) 
# to left of pivot and all greater elements to right 
# of pivot 
def partition(arr,low,high): 
    smaller_end_index = ( low - 1 )         # index of smaller element 
    pivot = arr[high]     # pivot 
  
    for j in range(low , high): 
  
        # If current element is smaller than or 
        # equal to pivot 
        if arr[j] <= pivot: 
          
            # increment index of smaller element 
            smaller_end_index += 1 
            arr[smaller_end_index], arr[j] = arr[j], arr[smaller_end_index] 
  
    # swap pivot index to inbetween smaller and larger
    arr[smaller_end_index + 1], arr[high] = arr[high], arr[smaller_end_index + 1] 
    return ( smaller_end_index + 1 ) 
  
# The main function that implements QuickSort 
# arr[] --> Array to be sorted, 
# low  --> Starting index, 
# high  --> Ending index 
  
# Function to do Quick sort 
def quickSort(arr,low,high): 
    if low < high: 
  
        # pi is partitioning index, arr[p] is now 
        # at right place 
        pi = partition(arr,low,high) 
  
        # Separately sort elements before 
        # partition and after partition 
        quickSort(arr, low, pi - 1) 
        quickSort(arr, pi + 1, high) 

        
# Driver code to test above 
arr = [10, 7, 8, 9, 1, 5] 
n = len(arr) 
quickSort(arr,0,n-1) 
print ("Sorted array is:") 
for i in range(n): 
    print ("%d" %arr[i]), 
# This code is contributed by Mohit Kumra 
```

Complexity: n = elements, d = number of max digits (300 would be 3), b = base of number (usually 10)<br>
 Best / average: O(nd)<br>
 Worst: O(n+d)<br>

Assume 16 numbers to be sorted with 5 digits each:

Radix sort = 16 * 5 = 80 time units.<br>
Quick sort = 16 * 4 = 64 time units.   4 is log 16

## LCM

```python
def gcd(a, b):
    ans = a % b
    if ans == 0:               
        return b
    else:        
        return gcd(b, ans)


def lcm_naive(a, b):
    lcm = (a*b)//gcd(a,b)
    return lcm
```

```python
A = [3, 2, 5]
B = [5, 1, 2]
n = 3

product = [0] * (2*n-1)
print(product)

for i in range(0,n):
    for j in range(0,n):
        product[i+j] = product[i+j]+(A[i]*B[j])
print(product)

```
